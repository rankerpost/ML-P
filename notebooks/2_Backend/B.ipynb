# Portable models to calculate predictions anywhere

## Input data to predict

```python
form_data = {
    'Industry': 'Materials',
    'Ethnicity': 'White',
    'Gender': 'Female',
    'Age': 18,
    'CivilStatus': 'Married',
    'YearsEmployed': 10,
    'Income': 540
}
```

```python
import pandas as pd

name = 'Alice'
df_input = pd.DataFrame(form_data, index=[name])
df_input
```

## Load model from previous lesson

```python
import pickle

path = '../../artifacts/pipeline.pkl'

with open(path, 'rb') as file:
    pipeline = pickle.load(file)
```

## Calculate the prediction

```python
pipeline.predict(df_input)
```

```python
probs = pipeline.predict_proba(df_input)
probs
```

```python
prob = probs[0][1]
prob
```
